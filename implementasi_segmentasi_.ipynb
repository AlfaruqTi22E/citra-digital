```python

import imageio.v3 as imageio
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import sobel

# Load the image (assume grayscale)
image_path = "/content/drive/MyDrive/Colab Notebooks/dataset/image.jpg"
image = imageio.imread(image_path, pilmode="L")

# Apply Sobel edge detection
sobel_x = sobel(image, axis=0)  # Sobel filter in X direction
sobel_y = sobel(image, axis=1)  # Sobel filter in Y direction
sobel_edges = np.hypot(sobel_x, sobel_y)  # Combine Sobel X and Y

# Normalize the Sobel edges
sobel_edges = (sobel_edges / sobel_edges.max()) * 255
sobel_edges = sobel_edges.astype(np.uint8)

# Apply Basic Thresholding
threshold_value = 128  # You can adjust this value
binary_image = np.where(sobel_edges > threshold_value, 255, 0).astype(np.uint8)

# Plot the results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image, cmap="gray")
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(sobel_edges, cmap="gray")
axes[1].set_title("Sobel Edge Detection")
axes[1].axis("off")

axes[2].imshow(binary_image, cmap="gray")
axes[2].set_title(f"Thresholded Image (Threshold={threshold_value})")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
